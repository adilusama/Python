### Implementing Line Charts with Matplotlib

```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

# GDP per Capita Growth in USA & China 

```python
usa = data[data.country == 'United States']
```

```python
plt.plot (usa.year, usa.gdpPerCapita)
plt.title ('GDP per Capita Growth in USA')
plt.xlabel ('Year')
plt.ylabel ('GDP per Capita')
plt.show()
```

```python
china = data[data.country == 'China']
```

```python
plt.plot (china.year, china.gdpPerCapita)
plt.title ('GDP per Capita Growth in China')
plt.xlabel ('Year')
plt.ylabel ('GDP per Capita')
plt.show()
```

## Populating data for both countries on same chart 

```python
plt.plot (usa.year, usa.gdpPerCapita)
plt.plot (china.year, china.gdpPerCapita)
plt.title ('GDP Per Capita Growth')
plt.legend (['USA', 'China'])
plt.xlabel ('Year')
plt.ylabel ('GDP per Capita')
plt.show()
```
