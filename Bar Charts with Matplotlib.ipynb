# Comparing the populations of 10 most populous countries in 2007

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
data_2007 = data[data.year == 2007]
data_2007.head()
```

```python
top10 = data_2007.sort_values('population', ascending = False).head(10)
top10.head(10)
```

```python
x = range (10)
plt.bar (x, top10.population / 10**9)
plt.xticks (x, top10.country, rotation = 'vertical')
plt.title ('10 Most Populous Countries')
plt.ylabel ('Population in Millions')
plt.show()
```

# Comparing the GDP of 10 most populous countries in 2007

```python
top10 = data_2007.sort_values('population', ascending = False).head(10)
top10.head(10)
```

```python
plt.subplot (2,1,1)
x = range (10)
plt.bar(x, top10.population / 10**6)
plt.xticks ([],[])
plt.title ('10 Most Populous Countries')
plt.legend (['Population in Millions'])

plt.subplot (2,1,2)
plt.bar (x, top10.gdpPerCapita * top10.population / 10**9)
plt.xticks (x, top10.country, rotation = 'vertical')
plt.legend (['GDP in Billions'])
plt.show()
```
