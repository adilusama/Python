### Population Growth Trend in USA & China from Countries.csv

```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
usa = data[data.country == 'United States']
china = data[data.country == 'China']
```

```python
plt.title ('Population for both countries (in millions)')
plt.plot (usa.year, usa.population / 10**6)
plt.plot (china.year, china.population / 10**6)
plt.legend (['USA', 'China'])
plt.xlabel ('Year')
plt.ylabel ('Population')
plt.show ()
```

```python
usa_growth = usa.population / usa.population.iloc[0] * 100 
china_growth = china.population / china.population.iloc[0] * 100
```

```python
plt.title ('Population Growth for both Countries')
plt.plot (usa.year, usa_growth)
plt.plot (china.year, china_growth)
plt.legend (['USA', 'China'])
plt.xlabel ('year')
plt.ylabel ('Population Growth')
plt.show()
```
