# Relationship between GDP per Capita and Life Expectancy

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('countries.csv')
data.head ()
```

```python
data_2007 = data [data.year == 2007]
```

```python
data_2007.head()
```

```python
plt.scatter (data_2007.gdpPerCapita, data_2007.lifeExpectancy, 5)
plt.title ('GDP per Capita and Life Expectancy in 2007')
plt.xlabel ('GDP per Capita ($)')
plt.ylabel ('Life Expectancy')
plt.show()
```

```python
data_2007.gdpPerCapita.corr(data_2007.lifeExpectancy)
```

```python
np.log10(data_2007.gdpPerCapita).corr(data_2007.lifeExpectancy)
```

```python
year_sorted = sorted(set(data.year))
```

```python
for given_year in year_sorted:                                          #Presenting Scatter plots for each year
        data_year = data [data.year == given_year]                      #Showing the correlation between GDP and LE
        plt.scatter (data_year.gdpPerCapita, data_year.lifeExpectancy, 5)
        plt.xlim(0,60000)
        plt.ylim(25,85)
        plt.xlabel ('GDP per Capita ($)')
        plt.ylabel ('Life Expectancy')
        plt.title (given_year)
        plt.show()
        #plt.savefig (str(given_year), dpi = 200)
        #plt.clf() 
```

```python
for given_year in year_sorted:                                  #Presenting Scatter plots for each year
        data_year = data [data.year == given_year]              #Showing the correlation between log(10)GDP and LE
        plt.scatter (np.log10(data_year.gdpPerCapita), data_year.lifeExpectancy, 5)
        plt.xlim(2,5)
        plt.ylim(25,85)
        plt.xlabel ('GDP per Capita ($)')
        plt.ylabel ('Life Expectancy')
        plt.title (given_year)
        plt.show()
        #plt.savefig ('log_' + str(given_year), dpi = 200)
        #plt.clf() 
```
